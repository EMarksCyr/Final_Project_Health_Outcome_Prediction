```python
from splinter import Browser
from bs4 import BeautifulSoup as soup
from webdriver_manager.chrome import ChromeDriverManager
from collections import namedtuple
```

```python
# Set up Splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# Visit coding scheme page
url = 'https://www.cdc.gov/brfss/annual_data/2019/pdf/codebook19_llcp-v2-508.HTML'
browser.visit(url)
```

```python
# Parse the HTML
html = browser.html
html_soup = soup(html, 'html.parser')
```

```python
#Scrape SAS variable names, question numbers and question text
headers = html_soup.find_all('td', class_='l m linecontent')
tables_uncut =  html_soup.find_all('table', class_='table')

for tables in tables_uncut:
    table = tables.find('tbody')
```

```python

HeaderInfo = namedtuple('HeaderInfo', ['lable', 'name', 'question'])

def get_value(header_line):
    index = header_line.find(':')
    label = header_line[index + 1:].strip()
    return label

def get_header_info(header):
    var_label = get_value(header.contents[0])
    sas_var_names = get_value(header.contents[12])
    var_text = get_value(header.contents[16])
    return HeaderInfo(var_label, sas_var_names, var_text)

variable_info_list = []

for header in headers:
    variable_info_list.append(get_header_info(header))
    
    
```
